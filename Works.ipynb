```python
import pandas as pd
```

```python
# Load entire file
chunker = pd.read_csv("/home/pi/Downloads/works-20210226.csv", chunksize=10000)
works = pd.concat(chunker, ignore_index=True)
```

```python
# Make sure the date column is in datetime format
works['creation date'] = pd.to_datetime(works['creation date'])
```

```python
# Sort the DataFrame in ascending order
works.sort_values(by=['creation date'], inplace=True, ignore_index=True)

# Get rid of last column
works = works.iloc[:,:6]
```

```python
# Test: Selecting data for the year of 2008
works_2008 = works[works['creation date'].dt.year == 2008]
works_2008
```

```python
# Test: write to local file

path = 'works-2008.csv'
works_2008.to_csv(path)
```

```python
# Loop action for the year 2008 to 2021
for i in range(2008,2021+1):
    path = f"works-{i}.csv"
    works[works['creation date'].dt.year == i].to_csv(path, index=False)
```
